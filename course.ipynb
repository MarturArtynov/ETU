```python
size = data.shape # размер таблицы
print("Lines:", size[0])
print("Rows:", size[1])
```

```python
data.dtypes # типы данных в таблице
```

```python
data.describe() # статистическое описание
```

```python
# 2. Отсортировать данные по какому-либо признаку.
```

```python
data.sort_values(by = 'price', inplace = True) # отсортировали
data.tail(5) # последние 5 отсортированных строк
```

```python
# 3.Удалить столбец\строку в таблице.
```

```python
data.drop([1479], inplace = True) # удалим последнюю строку с индекс 1479
data.tail(5)
```

```python
# 4. Посчитать количество уникальных значений в таблице(столбце).
```

```python
print("Count of unique genres: ", data['prime_genre'].nunique())
```

```python
#5. Найти минимум и максимум некоторого признака.
```

```python
print("Minimum rating: ", data['user_rating'].min())
print("Maximum rating: ", data['user_rating'].max())
```

```python
# 6. Переименовать столбец.
```

```python
data.rename(columns = {'ver': 'version'}, inplace = True)
data.columns
```

```python
# 7. Используя loc(iloc) отфильтровать данные.
```

```python
data.loc[:, ['track_name', 'user_rating']][data["user_rating"] > 4].head(5)
```

```python
# 8. Сгруппировать данные по какому-либо признаку.
```

```python
data.groupby('prime_genre')['price'].mean().sort_values(ascending = False)
```

```python
# 9. Используя value counts посчитать некоторый признак.
```

```python
data['rating_count_tot'].value_counts()
```

```python
# 10. С помощью apply и некоторого условия(функции) создать новый признак(столбец).
```

```python
data['size_mbytes'] = data['size_bytes'].apply(lambda k: k / (1000 * 1000))
data.head(5)
```

```python
# 11. Построить гистограмму некоторого признака (если это возможно). Подписать график.
```

```python
import matplotlib.pyplot as pyplot

genres = data.groupby('prime_genre').mean()
genres['user_rating'].plot(kind = 'bar')

pyplot.xlabel('Genres')
pyplot.ylabel('Mean rating')
pyplot.show()
```

```python
# 12. Построить график распредения любого признака.
```

```python
data[data['price'] > 0]['price'].plot()

pyplot.xlabel("Count of apps")
pyplot.ylabel("Price")
pyplot.show()
```
